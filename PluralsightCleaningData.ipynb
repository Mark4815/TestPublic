```python
data_df.isnull().sum()/len(data_df)*100
#Determines % of values per column that are null. Possibility to remove columns altogether where this is too high (Company)
```

```python
data_df.drop('Company', inplace=True, axis=1)
#axis = 1 says that we want to remove columns.
```

```python
data_df.duplicated().sum()
#Outputs true if any specific row matches with any other row. By then wrapping a sum() round it, you are counting the number of duplicate rows.
```

```python
data_df.drop_duplicates(inplace=True, keep='first')
#this drops all duplicate rows, keeping the first iteration of the row flagged. Same with 'last' if you wanted to keep the last record found instead.
```

```python
data_df['reservation_status_date'] = pd.to_datetime(data_df['reservation_status_date'])
#changes the format of the specified column to what you want. In this case, datetime format.
```

```python
data_df['reservation_status_year'] = data_df['reservation_status_date'].dt.year
data_df.head()
#this is taking a column and adding a new column with it formatted in a certian way, you can also use month, day, week, dayofweek
```

```python
cat_cols = [col for col in data_df.columns if data_df[col].dtype=='object']
cat_cols
#this gives you a list comphrehension for the columns where the data type is object.
```

```python
cat_df = data_df[cat_cols]
cat_df.head()
#This takes the above and uses it against the df to create a revised set of data that are just objects.
```

```python
cat_df['arrival_date_month'] = cat_df['arrival_date_month'].map({'January':1, 'February':2, 'March':3, 'April':4, 'May':5})
cat_df.head(50)
```
